```python
import os, json
import pycuber as pc
from pycuber.solver import CFOPSolver
from utils.visualisation import visualize_scramble

data = []

path = "cfop-dataset/training.seq.0"     # numbers instead of colors (seq)
 
with open(path, "r") as file:
    lines = file.read().strip().splitlines()
    for i in range(0, len(lines), 2):
        cube_state = lines[i]
        solution = lines[i+1]
        data.append({"state": cube_state, "solution": solution})

print("data len:", len(data))
print("first element:", data[0])
random_state = data[0]["state"].split()    # needs to be an array
print("first state:", random_state)
print("-> type:", type(random_state))

visualize_scramble(random_state)

visualize_scramble(data[1]["state"].split())
```
